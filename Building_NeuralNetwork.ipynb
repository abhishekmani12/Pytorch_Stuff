IMPORTS:

```python
import torch
import torchvision
from torchvision import transforms, datasets
```

```python
import torch.nn as nn #for creating neural networks
import torch.nn.functional as F #for functions involved in creating nn
```

Dataset:

```python
train=datasets.MNIST('',train=True,download=True,
                     transform=transforms.Compose
                     ([transforms.ToTensor()]))

test=datasets.MNIST('',train=False,download=True,
                    transform=transforms.Compose
                    ([transforms.ToTensor()]))

trainset=torch.utils.data.DataLoader(train, batch_size=10,shuffle=True)

testset=torch.utils.data.DataLoader(test, batch_size=10,shuffle=True)
```

NET Class:

4 layers are defined here

```python
class Net(nn.Module):      #Net inherits from nn.module class
    def __init__(self):
        super().__init__() #running the init method of the parent class

        #self.layerName=nn.linear( inputSize, outputSize )
        
        self.l1=nn.Linear(28*28,64) # Layer 1: '28*28' is the size of the input images, 64 output connections will be produced.
                                    #'28*28' will be flattened to '1*784' for the network to accept

        self.l2=nn.Linear(64,64) # Layer 2: 64 in, 64 out

        self.l3=nn.Linear(64,64) # Layer 3: 64 in, 64 out

        self.l4=nn.Linear(64,10) # Layer 4: 64 in, 10 out as this is the final output layer and there are only 10 output classes for this dataset

        #fully connected and feed forward

    def forward(self,x):    #to specify the passage of data 'x'
                               
        x=F.relu(self.l1(x)) #Activation function defines the output of the neuron by taking the sum of input data * weights and determines to fire/not fire the neuron
                                 #'relu'-Rectified Linear activation function is used,it scales data b/w 0-1 
        x=F.relu(self.l2(x))

        x=F.relu(self.l3(x))

        x=self.l4(x)

        return F.log_softmax(x, dim=1) #softmax af is used for output layer to convert output values into a probability distribution for a multi class classification 

net=Net()
print(net)
```

Testing for Output:

```python
#creating a random 28*28 image

X=torch.randn((28,28))

X=X.view(-1,28*28) #'-1' denotes as an any size variable
```
